# Dependent Sample (Paired T-test)
Same group(Sci&Tech) but different condition(etest_p,mba_p)

```python
from scipy.stats import ttest_rel

etest =  dataset[dataset["degree_t"] == "Sci&Tech"]["etest_p"]
mba_p =  dataset[dataset["degree_t"] == "Sci&Tech"]["mba_p"]
ttest_rel(etest,mba_p)
#pvalue is not acceptable and we don't have similiarity between this two conditions. 
```

# Indepenent Sample (Unpaired T-test)
Different group(Commerce,Arts) but Same condition(degree_p)

```python
from scipy.stats import ttest_ind

Comm =  dataset[dataset["hsc_s"] == "Commerce"]["degree_p"]
Arts =  dataset[dataset["hsc_s"] == "Arts"]["degree_p"]
ttest_ind(Comm,Arts)
#Pvalue is within acceptable range(<0.05 0r 5%) and we can see similiarity between two groups
```

# Hypothesis

```python
#statement : Test whether commenerence degree and Arts degree is same?
```

```python
from scipy.stats import ttest_ind

Comm =  dataset[dataset["hsc_s"] == "Commerce"]["degree_p"]
Arts =  dataset[dataset["hsc_s"] == "Arts"]["degree_p"]
ttest_ind(Comm,Arts)
```

```python
# P< 0.05 should be 
#Null hypothesis is accept,Alternate hypothesis is rejected.
```

```python
import scipy.stats as stats
stats.f_oneway(dataset["ssc_p"],dataset["hsc_p"])
```

```python
#Pvalue < 0.05)
#I accept this,This is significant difference (H1) (True) 
#There is no significant difference (H0) (False) 
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('salary ~ ssc_p + hsc_p + ssc_p:hsc_p', data=dataset).fit()
result = sm.stats.anova_lm(model, type=2) 
print(result)
```
